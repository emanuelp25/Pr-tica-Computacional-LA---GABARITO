```python
#plotando o espectro médio
plt.plot(file.iloc[1000:, 0], file.iloc[1000:, 201])
plt.title('Espectro médio')
plt.show()
```

```python
#selecionando a região do laser aleatório
#plotando o espectro médio
plt.plot(file.iloc[1650:1900, 0], file.iloc[1650:1900, 201])
plt.show()
```

```python
espectro = file.iloc[1650:1900, 201]
comprimento_de_onda = file.iloc[1650:1900, 0]

# Encontrar o valor máximo (pico) do espectro
valor_pico = np.max(espectro)

# Calcular a metade da altura do pico
metade_altura_pico = valor_pico / 2

# Encontrar os índices onde o espectro atinge metade da altura do pico
indices_metade_altura = np.where(espectro >= metade_altura_pico)[0]

# Verificar se existem pontos onde o espectro atinge a metade da altura do pico
if len(indices_metade_altura) > 0:
    primeiro_indice_metade_altura = indices_metade_altura[0]
    ultimo_indice_metade_altura = indices_metade_altura[-1]

    # Calcular a largura na metade da altura do pico
    largura_na_metade_da_altura = comprimento_de_onda.iloc[ultimo_indice_metade_altura] - comprimento_de_onda.iloc[primeiro_indice_metade_altura]

    # Plotar o espectro
    plt.plot(comprimento_de_onda, espectro)
    plt.xlabel('Comprimento de Onda')
    plt.ylabel('Espectro')
    plt.title('Espectro')
    plt.axhline(y=metade_altura_pico, color='r', linestyle='--', label='Metade da Altura do Pico')
    plt.legend()
    plt.show()

    # Imprimir os resultados
    print(f"Valor do Pico: {valor_pico}")
    print(f"Largura na Metade da Altura do Pico: {largura_na_metade_da_altura}")
else:
    print("O espectro não atinge a metade da altura do pico.")
```
